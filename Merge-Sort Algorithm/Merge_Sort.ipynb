```python
def merge_sort(array):
    if len(array) <= 1:
        return array  # Return the array if it's already sorted

    middle_point = len(array) // 2
    left_part = merge_sort(array[:middle_point])  # Sort left part
    right_part = merge_sort(array[middle_point:])  # Sort right part

    # Merge sorted parts
    merged_array = []
    left_array_index = 0
    right_array_index = 0

    while left_array_index < len(left_part) and right_array_index < len(right_part):
        if left_part[left_array_index] < right_part[right_array_index]:
            merged_array.append(left_part[left_array_index])
            left_array_index += 1
        else:
            merged_array.append(right_part[right_array_index])
            right_array_index += 1

    # If there are remaining elements in left_part
    while left_array_index < len(left_part):
        merged_array.append(left_part[left_array_index])
        left_array_index += 1

    # If there are remaining elements in right_part
    while right_array_index < len(right_part):
        merged_array.append(right_part[right_array_index])
        right_array_index += 1

    return merged_array  # Return the merged array

if __name__ == '__main__':
    numbers = [4, 10, 6, 14, 2, 1, 8, 5]
    sorted_numbers = merge_sort(numbers)
    print('Sorted array: ')
    print(sorted_numbers)
```
