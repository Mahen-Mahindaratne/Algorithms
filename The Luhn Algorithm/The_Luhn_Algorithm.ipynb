```python
def verify_card_number(card_number):
    # Initialize the sum of the digits in odd positions
    sum_of_odd_digits = 0
    # Reverse the card number to facilitate processing
    card_number_reversed = card_number[::-1]
    # Extract digits in odd positions (originally even positions)
    odd_digits = card_number_reversed[::2]

    # Sum the digits in odd positions
    for digit in odd_digits:
        sum_of_odd_digits += int(digit)

    # Initialize the sum of the digits in even positions
    sum_of_even_digits = 0
    # Extract digits in even positions (originally odd positions)
    even_digits = card_number_reversed[1::2]

    # Process each even-positioned digit
    for digit in even_digits:
        number = int(digit) * 2  # Double the digit
        if number >= 10:  # If the result is two digits
            # Sum the digits of the result (e.g., for 12, add 1 + 2)
            number = (number // 10) + (number % 10)
        sum_of_even_digits += number  # Add to the sum of even digits

    # Total the sums of odd and even positioned digits
    total = sum_of_odd_digits + sum_of_even_digits
    print(total)  # Print the total for debugging purposes

    # Check if the total is divisible by 10 (validating the card number)
    return total % 10 == 0

def main():
    # Example credit card number to verify
    card_number = '4111-1111-4555-1142'
    # Create a translation table to remove hyphens and spaces
    card_translation = str.maketrans({'-': '', ' ': ''})
    # Translate the card number to remove any formatting
    translated_card_number = card_number.translate(card_translation)

    # Verify the card number and print the result
    if verify_card_number(translated_card_number):
        print('VALID!')  # Valid card number
    else:
        print('INVALID!')  # Invalid card number

# Run the main function
main()
```
